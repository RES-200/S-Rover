```python
import cv2 as cv
import sys
import numpy as np
import os
import glob
import re
import shutil


stream_source = "http://192.168.4.1/camera"
file_source = r"C:\Users\shwet\Desktop\yesist final\input videos\WhatsApp Video 2024-03-01 at 00.43.04_7fda4119.mp4" 

cap = cv.VideoCapture(stream_source)

if not cap.isOpened():
    print("Stream source not available. Switching to file source.")
    cap = cv.VideoCapture(file_source)

if not cap.isOpened():
    print("Error opening video source")
    exit()

frame_interval = 7  
frame_count_new = 0 

frames_dir = r'C:\Users\shwet\Desktop\yesist final\STREAM FRAMES'
if not os.path.exists(frames_dir):
    os.makedirs(frames_dir)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_count_new % frame_interval == 0:
        cv.imwrite(os.path.join(frames_dir, f'frame{frame_count_new}.jpg'), frame)  

    frame_count_new += 1

cap.release()
```

```python

image_dir = r'C:\Users\shwet\Desktop\yesist final\frames'   
annuli_dir =  r'C:\Users\shwet\Desktop\yesist final\annulus ring strips' 
strips_dir =  r'C:\Users\shwet\Desktop\yesist final\rectangular strips' 
rings_dir =   r'C:\Users\shwet\Desktop\yesist final\mapped rings'
destination_dir = r'C:\Users\shwet\Desktop\yesist final\color_maps'
source_dir = os.getcwd()
image_paths = glob.glob(os.path.join(image_dir, '*'))
image_paths.sort(key=lambda f: int(re.sub('\D', '', f)))
grid_spacing = 50  # Adjust as needed
grid_color = (0, 255, 0)  #(BGR format)
strips = []

def extract_circular_strip(image, center, inner_radius, outer_radius):
    mask = np.zeros_like(image)
    cv.circle(mask, center, outer_radius, (255,255,255), -1)  
    cv.circle(mask, center, inner_radius, (0,0,0), -1) 
    result = cv.bitwise_and(image, mask)

    return result

center = (310,128)  # x, y (310,128) || (390,518)
inner_radius = 220  #220 || 502
outer_radius = 232 #232|| 518

for i, img_path in enumerate(image_paths):
    img = cv.imread(img_path)

    if img is None:
        print(f"Could not read the image: {img_path}")
        continue

    edges = cv.Canny(img,50,150)

    # contours
    contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

    # Drawing contours
    imgcontour = cv.drawContours(img, contours, -1, (0,0,0), 1)

    # circular strip
    strip = extract_circular_strip(imgcontour, center, inner_radius, outer_radius)

    # annulus
    cv.imwrite(os.path.join(annuli_dir, f'annulus{i}.jpg'), strip)

    img_rings = cv.circle(img.copy(), center, outer_radius, (0,255,0), 2)
    img_rings = cv.circle(img_rings, center, inner_radius, (0,255,0), 2)

    cv.imwrite(os.path.join(rings_dir, f'rings{i}.jpg'), img_rings)

    polar_image = cv.warpPolar(strip, dsize=(500,1570), center=center, maxRadius=outer_radius, flags=cv.WARP_POLAR_LINEAR)
    polar_image = cv.rotate(polar_image, cv.ROTATE_90_COUNTERCLOCKWISE)

    crop = polar_image[0:outer_radius-inner_radius,:]

    cv.imwrite(os.path.join(strips_dir, f'strip{i}.jpg'), crop)

    strips.append(crop)

    #adjust as needed
    inner_radius -= 4 # ||5
    outer_radius -= 4 # ||5

map = np.concatenate(strips, axis=0)

colormap1 = cv.applyColorMap(map, cv.COLORMAP_JET)
cv.imwrite("colored_map1.jpg", colormap1)
colormap2 = cv.applyColorMap(map, cv.COLORMAP_HOT)
cv.imwrite("colored_map2.jpg", colormap2)
colormap3 = cv.applyColorMap(map, cv.COLORMAP_PARULA)
cv.imwrite("colored_map3.jpg", colormap3)
colormap4 = cv.applyColorMap(map, cv.COLORMAP_TWILIGHT)
cv.imwrite("colored_map4.jpg", colormap4)
colormap5 = cv.applyColorMap(map, cv.COLORMAP_TWILIGHT_SHIFTED)
cv.imwrite("colored_map5.jpg", colormap5)
colormap6 = cv.applyColorMap(map, cv.COLORMAP_TURBO)
cv.imwrite("colored_map6.jpg", colormap6)
colormap7 = cv.applyColorMap(map, cv.COLORMAP_MAGMA)
cv.imwrite("colored_map7.jpg", colormap7)
colormap8 = cv.applyColorMap(map, cv.COLORMAP_INFERNO)
cv.imwrite("colored_map8.jpg", colormap8)
colormap9 = cv.applyColorMap(map, cv.COLORMAP_PLASMA)
cv.imwrite("colored_map9.jpg", colormap9)
colormap10 = cv.applyColorMap(map, cv.COLORMAP_VIRIDIS)
cv.imwrite("colored_map10.jpg", colormap10)
colormap11 = cv.applyColorMap(map, cv.COLORMAP_CIVIDIS)
cv.imwrite("colored_map11.jpg", colormap11)
colormap12 = cv.applyColorMap(map, cv.COLORMAP_COOL)
cv.imwrite("colored_map12.jpg", colormap12)
colormap13 = cv.applyColorMap(map, cv.COLORMAP_SPRING)
cv.imwrite("colored_map13.jpg", colormap13)
colormap14 = cv.applyColorMap(map, cv.COLORMAP_SUMMER)
cv.imwrite("colored_map14.jpg", colormap14)
colormap15 = cv.applyColorMap(map, cv.COLORMAP_AUTUMN)
cv.imwrite("colored_map15.jpg", colormap15)
colormap16 = cv.applyColorMap(map, cv.COLORMAP_WINTER)
cv.imwrite("colored_map16.jpg", colormap16)
colormap17 = cv.applyColorMap(map, cv.COLORMAP_BONE)
cv.imwrite("colored_map17.jpg", colormap17)
colormap18 = cv.applyColorMap(map, cv.COLORMAP_HSV)
cv.imwrite("colored_map18.jpg", colormap18)
colormap19 = cv.applyColorMap(map, cv.COLORMAP_RAINBOW)
cv.imwrite("colored_map19.jpg", colormap19)
colormap20 = cv.applyColorMap(map, cv.COLORMAP_OCEAN)
cv.imwrite("colored_map20.jpg", colormap20)

for i in range(1, 21):
    filename = f'colored_map{i}.jpg'
    source_file = os.path.join(source_dir, filename)
    destination_file = os.path.join(destination_dir, filename)
    shutil.move(source_file, destination_file)


cv.imwrite(r'C:\Users\shwet\Desktop\yesist final\maps\uncolored_map.jpg', map)
edges = cv.Canny(map,50,150)

contours, hierarchy = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
h, w, _ = map.shape
rows = h // grid_spacing
cols = w // grid_spacing
for x in np.linspace(start=0, stop=w, num=cols + 1):
    x = int(round(x))
    cv.line(map, (x, 0), (x, h), color=grid_color, thickness=1)
    cv.putText(map, str(x), (x, 20), cv.FONT_HERSHEY_SIMPLEX, 0.5, grid_color, 1, cv.LINE_AA)
for y in np.linspace(start=0, stop=h, num=rows + 1):
    y = int(round(y))
    cv.line(map, (0, y), (w, y), color=grid_color, thickness=1)
    cv.putText(map, str(y), (10, y), cv.FONT_HERSHEY_SIMPLEX, 0.5, grid_color, 1, cv.LINE_AA)
cv.imwrite(r'C:\Users\shwet\Desktop\yesist final\maps\map_with_contours_and_grid.jpg', map)
```
